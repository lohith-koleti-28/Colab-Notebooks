## Linear Regression

```python
import numpy as np
```

```python
inputs = np.array([[73, 67, 43],
                   [91, 88, 64],
                   [87, 134, 58],
                   [102, 43, 37],
                   [69, 96, 70]], dtype='float32')

targets = np.array([[56, 70],
                    [81, 101],
                    [119, 133],
                    [22, 37],
                    [103, 119]], dtype='float32')
```

```python
  inputs = torch.from_numpy(inputs)
  targets = torch.from_numpy(targets)
  print(inputs.shape)
  print(targets.shape)
```

```python
w = torch.randn(2,3,requires_grad=True)
b = torch.randn(2,requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
  return x @ w.t() + b
```

```python
preds = model(inputs)
print(preds)
print(targets)
print((preds - targets) ** 2)
```

```python
def mse(t1, t2):
  diff = t1 - t2
  return torch.sum(diff ** 2)/diff.numel()
```

```python
loss = mse(preds,targets)
```

```python
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

```python
preds = model(inputs)
preds
```

```python
loss = mse(preds,targets)
print(loss)
```

```python
loss.backward()
print(w.grad)
print(b.grad)
```

```python
with torch.no_grad():
  w -= w.grad * 1e-5
  b -= b.grad * 1e-5
  w.grad.zero_()
  b.grad.zero_()
```

```python
print(w)
print(b)
```

```python
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

```python
for i in range(10000):
  preds = model(inputs)
  loss = mse(preds, targets)
  loss.backward()
  with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
  if (i+1)%1000 == 0:
    print(f"Loss: {loss} for i: {i+1}")
```

```python
print(w)
print(b)
```

## Linear Regression using Python builtin functions

```python
import torch.nn as nn
import numpy as np
import torch
```

```python
inputs = np.array([[73, 67, 43],
                   [91, 88, 64],
                   [87, 134, 58],
                   [102, 43, 37],
                   [69, 96, 70],
                   [74, 66, 43],
                   [91, 87, 65],
                   [88, 134, 59],
                   [101, 44, 37],
                   [68, 96, 71],
                   [73, 66, 44],
                   [92, 87, 64],
                   [87, 135, 57],
                   [103, 43, 36],
                   [68, 97, 70]],
                  dtype='float32')

targets = np.array([[56, 70],
                    [81, 101],
                    [119, 133],
                    [22, 37],
                    [103, 119],
                    [57, 69],
                    [80, 102],
                    [118, 132],
                    [21, 38],
                    [104, 118],
                    [57, 69],
                    [82, 100],
                    [118, 134],
                    [20, 38],
                    [102, 120]],
                   dtype='float32')

inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
```

```python
from torch.utils.data import TensorDataset
```

```python
train_ds = TensorDataset(inputs, targets)
print(train_ds[0:3][1])
train_ds[0:3]
```

```python
from torch.utils.data import DataLoader
```

```python
batch_size = 5
train_dl = DataLoader(train_ds, batch_size, shuffle=True)
train_dl
```

```python
for xb, yb in train_dl:
  print(xb)
  print(yb)
```

```python
model = nn.Linear(3,2)
print(model.weight)
print(model.bias)
```

```python
list(model.parameters())
```

```python
preds = model(inputs)
preds
```

```python
import torch.nn.functional as F
```

```python
loss_fn = F.mse_loss
```

```python
loss = loss_fn(model(inputs), targets)
loss
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-6)
```

```python
def fit(num_epochs, model, loss_fn, opt, train_dl):
    for epoch in range(num_epochs):
        for xb, yb in train_dl:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            opt.step()
            opt.zero_grad()

        if (epoch+1) % (num_epochs / 10) == 0:
            print(f"Epoch: {epoch+1}/{num_epochs}, Loss: {loss:.4f}")
```

```python
fit(10000, model, loss_fn, optim, train_dl)
```
